####  Correlations

```python
# Correlation Function
# We are only interested in correlations greater than 0.9
def corr(df, column_name = ''):
    print (str(column_name) + ' correlations higher than 0.9')
    for i in df.columns[2:]:
        corr, p = pearsonr(df[column_name], df[i])
        if corr > 0.9:
            print (i)
```

#### Models
Wins and Losses

```python
#Function that returns the R value of after performing linear regression
def scr(x_values, y_values):
    x_train, x_test, y_train, y_test = train_test_split(x_values, y_values,train_size = 0.80, test_size = 0.20)
    lr = LinearRegression()
    lr.fit(x_train, y_train)
    y_predict = lr.predict(x_test)
    R = lr.score(x_train, y_train)
    return (R)
```

```python
# Independent variables with R^2 score of greater than 0.8 for 'Wins'
print ('Outcome: Wins Models')
for i in df.columns[2:20]:
    sc = scr(df[[i]],df[['Wins']])
    if sc > 0.8:
        print (i)
        print (sc)
```

```python
# Independent variables with R^2 score of greater than 0.8 for 'Losses'
print ('Outcome: Losses Models')
for i in df.columns[2:20]:
    sc = scr(df[[i]],df[['Losses']])
    if sc > 0.8:
        print (i)
        print (sc)
```
