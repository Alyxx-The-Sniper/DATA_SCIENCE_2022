```python
nba_2010 = nba[nba.year_id == 2010]
nba_2014 = nba[nba.year_id == 2014]
```

### `2010`
`Knicks pts VS. Nets pts per game` 

```python
knick_pts = nba_2010[nba_2010.fran_id == 'Knicks'].pts
nets_pts  = nba_2010[nba_2010.fran_id == 'Nets'].pts
```

### `game_location vs game_result`

```python
nba_2010[['game_location','game_result']].head()
```

### `Contingency tables of frequencies`

```python
# crosstab
location_result_freq = pd.crosstab(nba_2010.game_location, nba_2010.game_result)
location_result_freq
```

`1. 120 team won when they at homecourt advantage.
 2. 133 team loss when they are away.`

### `Contingency tables of proportions`

```python
location_result_prop = location_result_freq / len(nba_2010)
location_result_prop
```

`1. 29.5% of the team lose if their are away (A)
2. 26.6% of the team win their game when they are at homecourt (H)`

### `Marginal proportions`

```python
game_result_margin = location_result_prop.sum(axis=0)
game_result_margin
```

`1. There are 52.8% record of  losses
2. There are 47.0% record of  wins`

```python
game_location_margin = location_result_prop.sum(axis=1)
game_location_margin
```

`1. Total games by location(home/way) are distributed equally`

### `Expected contingency tables`

The more that the expected and observed tables differ, the more sure we can be that the variables are associated

```python
# expected
chi2, pval, dof, expected = chi2_contingency(location_result_freq)
expected
```

```python
# observed
location_result_freq
```

`Comparing our observed and expected result. 119 and 133, 106 abd 120 this values are at large differ from each other thus we can say that game_location has a relation to winning and lossing.`
