## Univariate Analysis

##### <font color='red'>1. What do coach ticket prices look like? What are the high and low values? What would be considered the average? Does $500 seem like a good price for a coach ticket?</font>

````python
sns.histplot(flight.coach_price)
# plt.hist(flight.coach_price)
plt.title('coach_price distribution')

plt.axvline(500, color='r', linestyle='dashed', label= 'coach_price')
plt.axvline(flight.coach_price.mean(), color='y', linestyle='dashed', label= 'coach_price average')
plt.legend()
plt.show()
plt.clf()
````

#### `Conclusion:` 
500$ seems a little bit pricey for the coach_price

````python
print(np.min(flight.coach_price))
print(np.mean(flight.coach_price))
print(np.max(flight.coach_price))
````

#### <font color='red'>2. Now visualize the coach ticket prices for flights that are 8 hours long. What are the high, low, and average prices for 8-hour-long flights? Does a $500 dollar ticket seem more reasonable than before?</font>


````python
## Task 2
coach_8_hours = flight[flight['hours'] == 8]
# coach_8_hours.describe(include= [np.number])
````

````python
sns.histplot(coach_8_hours.coach_price)
plt.title('8 hours coach_price distribution')
plt.axvline(coach_8_hours.coach_price.mean(), color='y', linestyle='dashed', label= 'coach_price 8hrs avg')
plt.legend()
plt.show()
plt.clf()
````

#### `conclusion:`

Still pricey, coach_price must be on average or lower 

````python
print(np.min(coach_8_hours.coach_price))
print(np.mean(coach_8_hours.coach_price))
print(np.max(coach_8_hours.coach_price))
````

#### <font color='red'>3. How are flight delay times distributed? Let's say there is a short amount of time between two connecting flights, and a flight delay would put the client at risk of missing their connecting flight. You want to better understand how often there are large delays so you can correctly set up connecting flights. What kinds of delays are typical?</font>

````python
## Task 3
sns.histplot(flight.delay, bins=4)
plt.show()
plt.clf()
````

````python
#Our histogram seems heavily skewed to the right
plt.figure(figsize=(12, 8)) 
sns.boxplot(x='delay', data=flight)
plt.show()
plt.clf()
````

````python
# check value_count for each unique value in delays
# There's seems too much variation and alot of outliers
# pd.set_option('display.max_rows', None)  # or 1000
flight.delay.value_counts().tail(10)
# theres a lot of 1 time occuring heavy delays
````

````python
# Cut off first the outliers in our histogram to make a better feel in our data
# now we can set our histogram to capture only delays that are lower than 200

sns.histplot(flight.delay[flight.delay <= 200], bins=10)
plt.show()
plt.clf()
````

````python
# most common delays are at 10minutes
flight.delay.mode()
````

#### `conclusion:`

Most common connecting flight delays are at 10 - 20mins

but when there's unexpeted happening's, delays can go around 1400 - 1500mins.


## Bivariate Analysis

#### <font color='red'>4. Create a visualization that shows the relationship between coach and first-class prices. What is the relationship between these two prices? Do flights with higher coach prices always have higher first-class prices as well?</font>

````python
## Task 4
sns.scatterplot(flight.coach_price, flight.firstclass_price)
plt.show()
plt.clf()
````

````python
# We have a large amount of entries in our scatter plot
# But I will not use a random sample, this data is fine with me

# Note: Since it took years for computer to load this huge entries, I will just use a sample size of 10%.. ehehe sowie ;(

perc = .10 # sample 10%
flight_sub = flight.sample(n = int(flight.shape[0]*perc))

sns.lmplot(x = "coach_price", y = "firstclass_price", data = flight_sub, line_kws={'color': 'black'}, lowess=True)
plt.show()
plt.clf()
````

#### `conclusion:`

Coach price and firstclass price are correlated and this means that the higher the value the higher the other value will get.

#### <font color='red'>5. What is the relationship between coach prices and inflight features &mdash; inflight meal, inflight entertainment, and inflight WiFi? Which features are associated with the highest increase in price?</font>

````python
## Task 5
sns.histplot(x= flight.coach_price, hue=flight.inflight_meal)
plt.show()
plt.clf()

sns.histplot(x= flight.coach_price, hue=flight.inflight_entertainment)
plt.show()
plt.clf()

sns.histplot(x= flight.coach_price, hue=flight.inflight_wifi)
plt.show()
plt.clf()

````

#### `conclusion:`

Majority of the flights that has meal,entertaiment and wifi has a higher coach_price

Inflight WIFI is the most associated with higher coach price. No wifi can means lower coach price
