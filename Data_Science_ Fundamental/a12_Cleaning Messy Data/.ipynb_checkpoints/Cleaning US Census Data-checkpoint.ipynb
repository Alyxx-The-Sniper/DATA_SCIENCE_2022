### 2. It will be easier to inspect this data once we have it in a DataFrame. You can't even call `.head()` on these `csv`s! How are you supposed to read them?

###   Using `glob`, loop through the census files available and load them into DataFrames. Then, concatenate all of those DataFrames together into one DataFrame, called something like `us_census`.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import seaborn as sns
````

````python
files = glob.glob('states*.csv')

df_list = []
for filename in files:
    data = pd.read_csv(filename)
    df_list.append(data)

# checking the number of items in our list, if we got all the csv's    
print(len(df_list))

us_census = pd.concat(df_list)
````

````python
# checking the total len of all csv's row combine together
sum(map(len, df_list)) 
````

## Regex to the Rescue

### 5. Use regex to turn the `Income` column into a format that is ready for conversion into a numerical type.

````python
us_census.Income = us_census['Income'].replace('\$', '', regex=True)
us_census.head(1)
````

### 6. Look at the `GenderPop` column. We are going to want to separate this into two columns, the `Men` column, and the `Women` column.

###   Split the column into those two new columns using `str.split` and separating out those results.

````python
GenderPop_split = us_census.GenderPop.str.split('_')
GenderPop_split.head(1)
````

````python
us_census['Men'] = GenderPop_split.str.get(0)
us_census['Women'] = GenderPop_split.str.get(1)
us_census = us_census.drop('GenderPop', axis=1)
us_census.head(3)
````

### 7. Convert both of the columns into numerical datatypes.

###   There is still an `M` or an `F` character in each entry! We should remove those before we convert.

````python
us_census.Men = us_census['Men'].replace('M', '', regex = True)
us_census.Women = us_census['Women'].replace('F', '', regex = True)
us_census.head(3)
````

````python
us_census.Men = pd.to_numeric(us_census.Men)
us_census.Women = pd.to_numeric(us_census.Women)
us_census.info()
````

### 9. You want to double check your work. You know from experience that these monstrous csv files probably have `nan` values in them! Print out your column with the number of women per state to see.

###    We can fill in those `nan`s by using pandas' `.fillna()` function.
   
###   You have the `TotalPop` per state, and you have the `Men` per state. As an estimate for the `nan` values in the `Women` column, you could use the `TotalPop` of that state minus the `Men` for that state.
   
   Print out the `Women` column after filling the `nan` values to see if it worked!

````python
# Check nan value using .info()
us_census.info()
# Under Women there are 57 non-null value while it should have 60
# Under Pacific there are 55 non-null value while it should have 60
````

````python
# View all rows with Nan values in the entire dataframe
us_census[us_census.isna().any(axis=1)]
````

````python
# We are only interested in Women column, lets view its rows with nan values
us_census[us_census['Women'].isna()]
````

````python
# replace fill this nan values with the diffrence of TotalPop minus Men population
us_census.Women = us_census['Women'].fillna(us_census.TotalPop - us_census.Men)

# Check again for nan in Women
us_census[us_census['Women'].isna()]
# We are now clear with the Nan values in Women
````

````python
# check again using .info()
us_census.info()
# our women column has now all the details we can start analysing it.
````

### 10. We forgot to check for duplicates! Use `.duplicated()` on your `census` DataFrame to see if we have duplicate rows in there.

````python
# Check for duplicates
# Remember that have Unnamed: 0 column, we dont want that to include in checking duplicates
# to exlude that we will need to use a subset 
duplicates = us_census.duplicated(subset = us_census.columns[1:])
duplicates.value_counts()
# We have nine duplicates
````

````python
# lets check where this duplicates
us_census_duploc = np.where(us_census.duplicated(subset = us_census.columns[1:]))
us_census_duploc
````

````python
# More Checking
dup_array = [ 6, 12, 18, 24, 30, 36, 42, 48, 54]

dup_list = []
for i in dup_array:
    location = us_census.iloc[i-1 : i+1]
    dup_list.append(location)

all_duplicates = pd.concat(dup_list)  
all_duplicates
````

### 11. Drop those duplicates using the `.drop_duplicates()` function.

````python
# We exclude the first column in our drop search
us_census = us_census.drop_duplicates(subset = us_census.columns[1:])
us_census.shape
# previously we had 60 row then after dropping 9 duplicates we have now 51 rows.
````

````python
# Check for Duplicates again.
duplicates = us_census.duplicated(subset = us_census.columns[1:])
duplicates.value_counts()
# We are now clear with duplicates 
````
