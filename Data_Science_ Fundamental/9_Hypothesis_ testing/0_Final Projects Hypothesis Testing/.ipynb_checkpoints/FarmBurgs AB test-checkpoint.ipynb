Farm life simulation

# II. Binomial Test  (Group B)
Test a sample frequency by a hypothetical population. 

`Null:` Number of purchases is equal or greater than 183

`Alt: ` Number of purchases is lower than 183



#### Creating a simulation to test out hypothesis

```python
null_outcomes = []
def simulation(observed_purchase_rate): 
    # We use random.choise between 'y' and 'n' with observed_purchase_rate (11.0%) probability of getting a 'y'.
    # our sample size is 1666 (Total number of group B population)
    # repeat this 10000 times
    for i in range(10000):
        simulated_visitors = np.random.choice(['y', 'n'], size=1666, p=[observed_purchase_rate, 1 - observed_purchase_rate])
        num_purchased = np.sum(simulated_visitors == 'y')
        null_outcomes.append(num_purchased)

    # Showing the first 10 results
    return null_outcomes[0:10]
```

```python
# input observed purchase rate, Data can view using contingency table above
simulation(0.11)
```

#### Inspecting the Null Distribution

As long as our observed purchase don't fall below the lower bound percentile, we are safely confident to achieve our target goal \$1000 per week.

Let's say:

- y = Sales of group C
- A = 169 (13% percentile)
- B = Group C price point

y = A * B
- `336.31`  = 169 * 1.99

Total population is 3 times larger the group C

- 3y = x
- 3(336.31) = x 
- `1008.93` = x 

We can achieve the $1000 target goal by maintaining the purchases equal or greater than the 13% percentile (169 purchases).

```python
# Setting to 13% percentile, one sided (greater)
# 87% Confident Interval
np.percentile(null_outcomes, [13,0])
```

```python
#
plt.figure(figsize=(9,5))
plt.hist(null_outcomes)
plt.axvline(183, color = 'g', linestyle = '--', label = 'Observed number of purchase')
plt.axvline(169, color = 'r', linestyle = '--', label = '13% Percentile')
# plt.axvline(138, color = 'r', linestyle = '--', label = 'Percentile higher bound')
plt.legend()
plt.show()
```

#### P value
We are only interested in getting the Pval of observing equal or greater than 13% percentile (169 purchases).

Our confidence internaval are at 87% that we can achieve our goal.

```python
# I encoded 169 instead of 183, it's because thats the minimum safe number of purchases we should get.
null_outcomes = np.array(null_outcomes)
pval_oneside = np.sum(null_outcomes >= 169) / len(null_outcomes)
pval_oneside
```

```python
# Let's try observing equal or greater than 183 purchase
# We are at 52% confident at this point
pval_oneside = np.sum(null_outcomes >= 183) / len(null_outcomes)
pval_oneside
```

#### Using the binom_test function

```python
pval_oneside_binom = binom_test(169, n=1666, p=0.11, alternative = 'greater')
str('{:.4f}'.format(pval_oneside_binom))
```

```python
pval_oneside_binom = binom_test(183, n=1666, p=0.11, alternative = 'greater')
str('{:.4f}'.format(pval_oneside_binom))
```

Results using the binom_test function and our manual code are similary the same.

## Conclusion

Failed to reject the null hypothesis.
At 52% confident, the number of purchases are equal or greater than 183.

# A and B test 
Or should we say B and C test? ;)

Testing Group B and C for significant differences.

`Null:` There is no significant difference from Group B and C in terms of sales revenue

`Alt: ` There is a significant difference from Group B and C in terms of sales revenue


Our baseline will be group B's succesful purchases and target goal is Group C's  revenue. We want to conduct our A/B test base on groups revenue not the purchase frequency. 

####  Setting Variables

```python
# Group B succesful purchase
purchase_199 = np.sum((farm.group == 'B') & (farm.is_purchase == 'Yes'))

# Group B total visits
sample_size = np.sum((farm.group == 'B')) 

# Group C succesful revenue
group_C_Revenue = 83 * 4.99

# Group B's purchases need to reach Group C' revenue
increase_target = group_C_Revenue / 1.99
```

#### Baseline

```python
baseline = (purchase_199 / sample_size) 
baseline
```

#### Minimum Detectable Effect




```python
# purchase rate to reach group C's revenue
increase_target_pct = increase_target / sample_size
increase_target_pct
```

```python
# percentage increase from baseline
lift = (increase_target_pct - baseline) / baseline 
lift
```

```python
significance_threshold = 0.05
results = []
results_pvals = []

for i in range(100):
    #simulatedata:
    group_B = np.random.choice(['yes','no'],size=int(sample_size/2),p=[baseline,1-baseline])
    group_C = np.random.choice(['yes','no'],size=int(sample_size/2),p=[increase_target_pct, 1-increase_target_pct])
    group=['group b']*int(sample_size/2)+['hypothetical group c']*int(sample_size/2)
    outcome=list(group_B)+list(group_C)
    sim_data={"GROUP":group,"IS_PURCHASED?":outcome}
    sim_data=pd.DataFrame(sim_data)

    #runthetest
    ab_contingency=pd.crosstab(np.array(sim_data.GROUP),np.array(sim_data['IS_PURCHASED?']))
    chi2,pval,dof,expected=chi2_contingency(ab_contingency)
    result=('significant'if pval < significance_threshold else 'not significant')
    result_pval = results_pvals.append(pval)

    # append the result to our results list here:
    results.append(result)
```

```python
# Showing first five p vals
results_pvals[0:5]
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='GROUP', hue='IS_PURCHASED?', data = sim_data)
```

```python
# Turn to np.array
results =  np.array(results)
```

```python
print("Proportion of significant results:")
# divide by 100 times (our `for loop` size)
print(np.sum(results == 'significant')/ 100 * 100) 
print("Proportion of NOT significant results:")
print(np.sum(results == 'not significant')/ 100 * 100)
```

# Conclusion

Group B and C are not significantly different. Either of the two will succesfully reach our targer goal, however Group B's price point is more attractive even though Group C has more revenue by a little in our initial test. When we run the test 100 times it shows that 87% results are NOT significant. In the long run Group B is more profitable.
