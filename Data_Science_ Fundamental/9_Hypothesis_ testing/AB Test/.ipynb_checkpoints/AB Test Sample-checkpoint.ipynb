### <font color='brown'>Baseline</font>

```python
baseline = (len(purchasing_customers) / len(customer_visits)) * 100
str(baseline)+'% of customers made an order' 
```

### <font color='brown'>Minimum Detectable Effect(Lift)</font>

Our lift price is 1240$

How many new customers we need to get this $1240 incerease?

The avg of money_spent of a customer is 26 thus we need 46 new customers to get this $1240 mark.

```python
# Frequency of customer
increase_target = 1240
avg_money_spent = np.mean(money_spent)
new_customer_needed = increase_target / avg_money_spent
print('Addiotnal customer needed: ' + str(round(new_customer_needed)))
```

```python
# proportion of customer
target_total_customer = len(purchasing_customers) + new_customer_needed
target_total_customer_pct = round(target_total_customer / len(customer_visits) * 100, 2)
print('Proportion: ' + str(target_total_customer_pct) + str('%'))
```

```python
# percentage increase from baseline
lift = (target_total_customer_pct - baseline) / baseline * 100
print('Our MDE/Lift:' + str(round(lift,2)))
```

 Wed like to be fairly certain, but this isnt going to be a million dollar decision, so lets go with 5%.

```python
Threshold = 0.05
```

## Using Power (100 loops)

```python
#presetvalues
significance_threshold = 0.05
sample_size = 630
lift = 50.22
baseline_rate = 18.6 / 100
mde_rate = target_total_customer_pct / 100
```

```python
results = []
pvals = []

for i in range(100):
    #simulatedata:
    nosh_old_design = np.random.choice(['yes','no'],size=int(sample_size/2),p=[baseline_rate,1-baseline_rate])
    nosh_new_design = np.random.choice(['yes','no'],size=int(sample_size/2),p=[mde_rate, 1-mde_rate])
    group=['old design']*int(sample_size/2)+['new design']*int(sample_size/2)
    outcome=list(nosh_old_design)+list(nosh_new_design)
    sim_data={"Design":group,"Purchased?":outcome}
    sim_data=pd.DataFrame(sim_data)

    #runthetest
    ab_contingency=pd.crosstab(np.array(sim_data.Design),np.array(sim_data['Purchased?']))
    chi2,pval,dof,expected=chi2_contingency(ab_contingency)
    result=('significant'if pval < significance_threshold else 'not significant')
    pvals.append(pval)
   
    # append the result to our results list here:
    results.append(result)
```

```python
# Showing first 5 results
print(results[0:5])
print(pvals[0:5])
```
