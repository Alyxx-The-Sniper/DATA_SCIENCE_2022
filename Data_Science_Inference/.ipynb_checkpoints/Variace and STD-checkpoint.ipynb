# Variance
Average of the sum of difference of each data point from the mean squared.


Formula: 


We do squared each differences to eliminate negative value. If we dont eliminate negatives we could run into problem summing up all the deffirences.

# Standard Deviation
Square root of variance

formula:


Variance is a tricky statistic to use because its units are different from the mean and the data itself(We aquared it to eliminate negatives).

We do square root the variance to revert back to its appropriate value, then we can compare this STD to the distribution mean. Standard Deviation is easier to interpret.

By finding the number of standard deviations a data point is away from the mean, we can begin to investigate how unusual that datapoint truly is. We can usually expect around 68% of your data to fall within one standard deviation of the mean, 95% of your data to fall within two standard deviations of the mean, and 99.7% of your data to fall within three standard deviations of the mean.



### Example 1:
Check the distribution of students grades per section.

```python
import numpy as np
# list of student grades per section
section_one = [80.24, 81.15, 81.29, 82.12, 82.52, 82.54, 82.76, 83.37, 83.42, 83.45, 83.47, 83.79, 83.91, 83.98, 84.03, 84.69, 84.74, 84.89, 84.95, 84.95, 85.02, 85.18, 85.53, 86.29, 86.83, 87.29, 87.47, 87.62, 88.04, 88.5]
section_ten = [65.82, 70.77, 71.46, 73.63, 74.62, 76.53, 76.86, 77.06, 78.46, 79.81, 80.64, 81.61, 81.84, 83.67, 84.44, 84.73, 84.74, 85.15, 86.55, 88.06, 88.53, 90.12, 91.27, 91.62, 92.86, 94.37, 95.64, 95.99, 97.69, 104.4]
```

```python
# Getting the mean
section_one_mean = np.mean(section_one)
section_ten_mean = np.mean(section_ten)

print(section_one_mean)
print(section_ten_mean)
```

Both sections has the same mean. Does section 1 and 10 students are equally smart in general?

We can't be certain by just taking the mean of the distribution. Let's inspect the spreadness of the distribution. 

```python
# Getting the standard deviation
section_one_std = np.std(section_one)
section_ten_std = np.std(section_ten)

print(section_one_std)
print(section_ten_std)
```

Though both sections have the same mean, their standard deviation is significant different.

```python
import matplotlib.pyplot as plt
# visualization
# Setting bin to 30, both sections has 30 students I want to plot each student.
plt.hist(section_one, alpha = 0.5, label = "Section one grades", bins=30)
plt.hist(section_ten, alpha = 0.4, label = "Section ten grades", bins=30)
plt.title("Student grades section 1 and 10")
plt.xlabel("Grades")
plt.legend()
plt.show()
```

`Interpretation`

Section ten distribution has more spread, this might enable some students to ace their tests, but other students get left behind while section one has compress distribution saying all students have a firm understanding of the subject.

### `Example 2`

Let's say we have a new student name 'Lisa' who's grade is 88. How well is Lisa's grade compare to section one and ten?

#### Investigating data points 
Section one disribution

```python
plt.figure(figsize=(10,5))

lisa_grade = 88
plt.axvline(lisa_grade, color='red', linestyle='-', label = "lisa's grade")

# Distribution of section one
plt.hist(section_one, alpha = 0.5, bins=30, edgecolor='black')
plt.title('Section one')
plt.xlabel('Grades')
plt.ylabel('Number of student')

# mean
plt.axvline(section_one_mean, color ='red', linestyle='--', label = "mean")

# one STD away from mean
plt.axvline(section_one_mean + section_one_std, color ='#AEE59C', linestyle='solid', label = "1 STD")
# two STD away from mean
plt.axvline(section_one_mean + section_one_std*2, color ='#82C66C', linestyle='solid', label = "2 STD")
# three STD away from mean
plt.axvline(section_one_mean + section_one_std*3, color = '#4C9235', linestyle='solid', label = "3 STD")

# negative STD away from mean
plt.axvline(section_one_mean - section_one_std, color = '#EACCA2', linestyle='solid', label = "-1 STD")
plt.axvline(section_one_mean - section_one_std*2, color = '#D1AE80', linestyle='solid', label = "-2 STD")
plt.axvline(section_one_mean - section_one_std*3, color = '#B38D5A', linestyle='solid', label = "-3 STD")

plt.legend()
```

Lisa's grade is 2 standard deviation away from the mean with respect to section one distribution.
We can say that Lisa outperform 68%(one std) of the students in section one.

#### Investigating data points 
Section ten disribution

```python
plt.figure(figsize=(10,5))

lisa_grade = 88
plt.axvline(lisa_grade, color='red', linestyle='-', label = "lisa's grade")

# Distribution of section one
plt.hist(section_ten, alpha = 0.5, bins=30, edgecolor='black') # edge color black to see seperation per observation
plt.title('Section one')
plt.xlabel('Grades')
plt.ylabel('Number ofstudent')

# mean
plt.axvline(section_ten_mean, color ='red', linestyle='--', label = "mean")

# one STD away from mean
plt.axvline(section_ten_mean + section_ten_std, color ='#AEE59C', linestyle='solid', label = "1 STD")
# two STD away from mean
plt.axvline(section_ten_mean + section_ten_std*2, color ='#82C66C', linestyle='solid', label = "2 STD")
# three STD away from mean
plt.axvline(section_ten_mean + section_ten_std*3, color = '#4C9235', linestyle='solid', label = "3 STD")

# negative STD away from mean
plt.axvline(section_ten_mean - section_ten_std, color = '#EACCA2', linestyle='solid', label = "-1 STD")
plt.axvline(section_ten_mean - section_ten_std*2, color = '#D1AE80', linestyle='solid', label = "-2 STD")
plt.axvline(section_ten_mean - section_ten_std*3, color = '#B38D5A', linestyle='solid', label = "-3 STD")

plt.legend()
```

Because the distribution of section ten is more spreadout (higher STD), lisa's grade falls only on 'one standard deviation' away from the mean. 

<font color='red'>Section ten students from 2 STD are way more better than section one!!.</font>
